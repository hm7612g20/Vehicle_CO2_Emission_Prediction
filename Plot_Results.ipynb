```python
import pandas as pd
import numpy as np

import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
model_specs = ["Simple Lin Reg: No. Cylinders", "Simple Lin Reg: Engine Size", "Simple Lin Reg: CFC (mpg)",
              "Simple Lin Reg: CFC", "Multi Lin Reg: CFC + No. Cylinders", "Multi Lin Reg: CFC + Engine Size",
              "Multi Lin Reg: CFC (mpg) + No. Cylinders", "Multi Lin Reg: CFC (mpg) + Engine Size",
              "Multi Lin Reg: CFC + Engine Size + Fuel Type", "Poly Reg (deg 2): CFC (mpg)", 
               "Poly Reg (deg 7): CFC (mpg)", "Poly Reg (deg 2): CFC (mpg) + Engine Size + Fuel Type"]
R2 = [0.693, 0.723, 0.823, 0.842, 0.873, 0.877, 0.886, 0.890, 0.991, 0.877, 0.881, 0.987]
MAE = [1051, 944, 605, 539, 434, 421, 390, 375, 30.4, 422, 407, 43.9]
MSE = [24.6, 23.3, 16.2, 14.1, 13.5, 13.5, 13.1, 12.7, 3.11, 10.8, 9.68, 4.44]
```

```python
# set width of bar 
barWidth = 0.25
#fig = plt.subplots(figsize =(12, 8)) 
 
fig, ax1 = plt.subplots()

# Set position of bar on X axis 
br1 = np.arange(len(model_specs)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

color1 = "tab:red"
ax1.set_xlabel("Model")
ax1.set_ylabel("R$^{2}$", color=color1)
ax1.bar(br1, R2, color=color1, width = barWidth, label="R$^{2}$")
ax1.tick_params(axis="y", labelcolor=color1)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color2 = "tab:blue"
color3 = "tab:green"
ax2.set_ylabel("Mean Absolute Error", color=color2)
ax2.bar(br2, MAE, color=color2, width = barWidth, label="MAE")
#ax2.bar(br3, MSE, color=color3, width = barWidth, label="MSE")
ax2.tick_params(axis="y", labelcolor=color2)

plt.grid(False)

plt.xticks([r + barWidth for r in range(len(model_specs))], model_specs, rotation=90)

#plt.legend()

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
# Figure Size
fig, ax = plt.subplots(figsize=(10, 15))
 
# Horizontal Bar Plot
color1 = "tab:red"
ax.barh(model_specs, R2, color=color1)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
ax.set_xlabel("R$^{2}$")
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Show top values 
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.01, i.get_y()+0.5, 
             str(i.get_width()),
             fontsize = 12,
             color=color1)
plt.show()
```
